```python
import pandas as pd
from datetime import datetime, date
```

```python
gvkey_df = pd.read_sas("./data/capitaliq_gvkey.sas7bdat")
gvkey_df['COMPANYID'] = gvkey_df['COMPANYID'].astype(int)
gvkey_df['GVKEY'] = gvkey_df.GVKEY.apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
gvkey_df['COMPANYNAME'] = gvkey_df.COMPANYNAME.apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
gvkey_df['STARTDATE'] = pd.to_datetime(gvkey_df['STARTDATE'], errors='coerce')
gvkey_df['ENDDATE'] = pd.to_datetime(gvkey_df['ENDDATE'], errors='coerce')

gvkey_df
```

```python
pd.read_sas("./data/compustat crsip link.sas7bdat")
```

```python
compustat_cols = ['GVKEY', 'TIC', 'LIID', 'LINKDT', 'LINKENDDT', 'LINKPRIM', 'LINKTYPE', 'LPERMCO', 'LPERMNO', 'CONM', 'CUSIP']
compustat_df = pd.read_sas("./data/compustat crsip link.sas7bdat")[compustat_cols]
# for col in compustat_df.columns:
#     print(col)
for col in compustat_df.columns:
    compustat_df[col] = compustat_df[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

compustat_df['LPERMNO'] = compustat_df['LPERMNO'].astype(int)
compustat_df.rename(columns={'LPERMNO':'PERMNO'}, inplace=True)
compustat_df['LINKDT'] = pd.to_datetime(compustat_df['LINKDT'])
compustat_df['LINKENDDT'] = pd.to_datetime(compustat_df['LINKENDDT'], errors='coerce')

compustat_df
```

```python
def mapping_valid(row):
    link_dt = row['LINKDT']
    start = row['STARTDATE']
    end = row['ENDDATE']
    # If both start and end are missing, assume always valid
    if pd.isna(start) and pd.isna(end):
        return True
    # If there is a start date, LINKDT must be on or after the start
    if pd.notna(start) and link_dt < start:
        return False
    # If there is an end date, LINKDT must be on or before the end
    if pd.notna(end) and link_dt > end:
        return False
    return True

df_merged = pd.merge(compustat_df, gvkey_df, on='GVKEY', how='left')
# Apply the validation function row-wise and filter the merged DataFrame.
df_merged['ValidMapping'] = df_merged.apply(mapping_valid, axis=1)
compustat_df_w_companyid = df_merged[df_merged['ValidMapping']].copy()
compustat_df_w_companyid .drop(columns=['ValidMapping'], inplace=True)

compustat_df_w_companyid 
```

```python
sp500_list = pd.read_stata("./data/MSP500LIST.dta")
value_counts = sp500_list.loc[~(sp500_list['ending'].dt.date < date(2008, 1, 1)), 'PERMNO'].value_counts()
value_counts[value_counts > 1]
sp500_list['PERMNO'] = sp500_list['PERMNO'].astype(int)
sp500_list['start'] = pd.to_datetime(sp500_list['start'])
sp500_list['ending'] = pd.to_datetime(sp500_list['ending'])
sp500_list
```

```python
current_date = pd.Timestamp(datetime.now().date())
# Merge left and right tables on PERMNO using a left join
merged_df = pd.merge(sp500_list, compustat_df_w_companyid , on='PERMNO', how='left')

# Define a helper function to get the effective end date
def effective_end(date, default=current_date):
    return date if pd.notnull(date) else default

# Filter rows where the periods overlap:
# Overlap exists if:
# left.start <= effective right.LINKENDDT AND effective left.ending >= right.LINKDT
merged_df = merged_df[
    (merged_df['start'] <= merged_df['LINKENDDT'].fillna(current_date)) &
    (merged_df['ending'].fillna(current_date) >= merged_df['LINKDT'])
]

sp500_merged_df = merged_df
sp500_merged_df = sp500_merged_df.dropna(subset=['COMPANYID'])
sp500_merged_df['COMPANYID'] = sp500_merged_df['COMPANYID'].astype(int)
sp500_merged_df['GVKEY'] = sp500_merged_df['GVKEY'].astype(str)
sp500_merged_df.to_csv("./sp500_list_w_gvkey_companyid.csv", index=False)
```
